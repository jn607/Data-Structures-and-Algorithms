Given two strings s and t, return true if t is an 
anagram
 of s, and false otherwise.

```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        def hash_creation(u):
            counter = {}    
            for char in u:
                if char not in counter:
                    counter[char] = 0
                counter[char] += 1
            return counter
        if hash_creation(s) == hash_creation(t):
            return True
        else:
            return False
```
